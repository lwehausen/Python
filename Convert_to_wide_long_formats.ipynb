Generate random data that is in standard format, then convert it to wide format.

```python
import pandas as pd
```

```python
#First, generate the code to create the dataframe that will be converted to wide format:
data_long = pd.DataFrame(columns = ["grp", "sex", "meanL", "sdL", "meanR", "sdR"])
data_long.loc[0] = ["A", "F", 0.225, 0.106, 0.340, 0.0849]
data_long.loc[1] = ["A", "M", 0.470, 0.325, 0.570, 0.3250]
data_long.loc[2] = ["B", "F", 0.325, 0.106, 0.400, 0.0707]
data_long.loc[3] = ["B", "M", 0.547, 0.308, 0.647, 0.2740]
data_long
```

```python
#Now generate the code to convert the original dataframe to wide format:
pd.pivot(data_long, index = "grp", columns = "sex", values = ["meanL", "sdL", "meanR", "sdR"])
```

Consider the pccc icd10 dataset.

```python
pc = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/pccc_icd10_dataset.csv")
```

Remove all the columns labeled with ”g” and a number.

```python
#First, let's take a look at the data:
pc.head(1)
```

```python
#Remove all the columns labeled with "g" and a number:
pc = pc.drop(columns=["g1", "g2", "g3", "g4", "g5", "g6", "g7", "g8", "g9", "g10"])
```

Now convert this to a long data set with three columns: id, type (pc or dx), and code.

```python
#Go from wide to narrow
pd.melt(pc, id_vars = "id",
        value_vars = ["dx1", "dx2", "dx3", "dx4", "dx5", "dx6", "dx7", "dx8", "dx9", "dx10", "pc1", "pc2", "pc3", "pc4","pc5", "pc6", "pc7", "pc8", "pc9", "pc10"],
        value_name = "code",
        var_name = "type")
```
