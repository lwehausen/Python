#Pie Charts

Proportions of different ethnic groups in the city of Chicago.

```python
#Data to plot
labels = ['White', 'Black', 'Hispanic', 'Asian', 'Other']
sizes = [32, 32, 28, 5, 3]
colors = ['pink', 'purple', 'yellow', 'blue', 'orange']

#Plotting the pie chart
plt.pie(sizes, labels = labels, colors= colors, autopct = '%.1f', shadow = True, startangle = 270)
plt.title('Chicago Diversity', fontweight= 'bold')
plt.axis('equal')
plt.show()
```

#Bar Graphs

Bar graph displaying number of wins from 2014-2018 of the Golden State Warriors.

```python
#Set x-axis labels and data
objects = ('2018', '2017', '2016', '2015', '2014')
y_pos = np.arange(len(objects))
performance = [58, 67, 73, 67, 51]

#Create bar plot with labels, a title, and color
plt.bar(y_pos, performance, align='center', alpha = 0.5, color = 'blue')

#Places the labels on the x-axis at their various locations determined by np.arange(len(objects))
plt.xticks(y_pos, objects, rotation = "horizontal")
plt.ylabel('Number of Wins')
plt.xlabel('Year')
plt.title('Golden State Warriors Wins')

plt.show()
```

#Double Bar Graphs

Double bar graph displaying percent of adolescents who play sports, by gender.

```python
#First set of data
Males = (40/100, 17/100, 20/100, 24/100, 8/100, 9/100, 5/100)
ind = np.arange(7) #the x locations for the groups
width = 0.4 #the width of the bars
fig, ax = plt.subplots()

#Creating the first set of bars for the first set of data
rects1=ax.bar(ind, Males, width, color = 'blue')

#Second set of data-- creation of bars
Females = (25/100, 17/100, 17/100, 17/100, 8/100, 12/100, 4/100)
rects2 = ax.bar(ind + width, Females, width, color='pink')

#Add some labels
ax.set_ylabel('Percent of Adolescents Who Play Sports', fontsize = 12)
ax.set_xlabel('Sports', fontsize = 12)
ax.set_title('Percent of Adolescent Male and Females Who Play Sports', fontsize = 13)
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('Basketball', 'Track', 'Soccer', 'Baseball/Softball', 'Tennis', 'Swimming', 'Cross Country'), fontsize = 7)

#Create a legend
ax.legend((rects1[0], rects2[1]), ('Males', 'Females'))

plt.savefig('Wehausen_Quiz2.jpg')

plt.show()
```

#Stacked Bar Graphs

Same data as previous graph that displays number of waterpark visitors.

```python
#Data for the first and second bar plots
children_visitors = (780, 1050, 3056, 5025)
adult_visitors = (315, 400, 1000, 1500)

#Setting the width of the bars and the tick positions
width = 0.5 # the width of the bars
ind = np.arange(1,5) # the x locations for the groups
tick_pos = [i + (width/50) for i in ind]

#Creating the bars
p1 = plt.bar(ind, children_visitors, width, color = 'pink', align = 'center')
p2 = plt.bar(ind, adult_visitors, width, bottom=children_visitors, color='purple',align='center')

#Add some text for labels, titles, axes ticks, and legend
plt.ylabel('Number of Visitors', fontsize=12)
plt.xlabel('Months', fontsize=12, labelpad=15)
plt.title('Number of Child and Adult Water Park Visits', fontsize=13)
plt.xticks(tick_pos, ('April', 'May', 'June', 'July'), fontsize=11)
plt.legend((p1[0], p2[1]), ('Chilren', 'Adults'), loc='best')

#Display the bar plot
plt.show()
```

#Contour Plots

General example of an unfilled contour map with cool and warm color mapping.

```python
#Creating the data
delta = 0.025
x = np.arange(0,4.0,delta)
y = np.arange(0,5.0,delta)

X, Y = np.meshgrid(x,y)

#Creates a Gaussian distribution for equal shape X, Y
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

plt.figure()

cp = plt.contour(X, Y, Z, cmap=cm.coolwarm)

#Labels and titles
plt.clabel(cp, inline=True, fontsize = 10)
plt.title('Countour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```

Same plot as the previous but filled and with no color.

```python
#Create the data
delta = 0.025
x = np.arange(0,4.0, delta)
y = np.arange(0, 5.0, delta)
#meshgrid makes rectangular arrays
X, Y = np.meshgrid(x,y)

#Creates Gaussian distributition
Z = np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)

plt.figure()

cp = plt.contourf(X, Y, Z)
plt.clabel(cp, colors = 'k', fmt = '%2.1f', fontsize = 12)
contour_filled = plt.contourf(X, Y, Z, cmap=cm.binary)

plt.colorbar(contour_filled)
plt.title('Filled Contour Plot')
plt.xlabel('X (cm)')
plt.ylabel('Y (cm)')
plt.show()
```
