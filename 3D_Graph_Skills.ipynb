#The purpose of this file is to display 3D graphing skills.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
```

3D heat plot of how neurons in primary visual cortex respond to images.

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.view_init(azim=60)

X=np.arange(-0.5, 6, 0.25)
Y=np.arange(-18, 2.8, 0.25)
X, Y = np.meshgrid (X,Y)
R= np.exp(-0.5 * (X**2/4 + Y**2/4)) * np.sin(2 * np.pi * 0.2 * X)
Z = np.exp(R)

surf= ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased = False)
ax.set_zlim(0.75, 2.6)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf,shrink=0.4, aspect=8, pad=0.08)

ax.set_xlabel("Simple Cell's Receptive Field in PVC", fontsize=6)
ax.set_ylabel('Visual Response Properties of Vertical Simple Cell', fontsize=6)
ax.set_zlabel('Neuron Response to Image', fontsize=6)
ax.set_title('How Neurons in Primary Visual Cortex Respond to Images', fontsize=10)
ax.dist = 13
plt.show()
```

General example of a cool-warm surface plot.

```python
#Creating the figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#Setting rotation angle of plot
ax.view_init(azim=45)

#Make data
X=np.arange(0, 4.0, 0.25)
Y=np.arange(0, 5.0, 0.25)
X, Y = np.meshgrid (X,Y)
R=  np.exp(-((X-1)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-3)**2+(Y-4)**2)/0.15) + \
    np.exp(-((X-2)**2+(Y-3)**2)/0.15) + \
    np.exp(-(X-2)**2) * np.exp(-(Y - ((X-2)**2+1))**2/0.15)
Z = np.exp(R)

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased= False)

ax.set_zlim(0.75, 2.6)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.6, aspect = 10, pad=0.09)

ax.set_xlabel('X axis', fontsize=10)
ax.set_ylabel('Y axis', fontsize=10)
ax.set_zlabel('Z axis', fontsize=10)
ax.set_title('Cool-Warm Surface Plot', fontsize=10)
ax.dist = 12
plt.show()
```

Same plot as previous but with Z-axis on opposite side.
