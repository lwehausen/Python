A data analysis and data visualization using the HELPrct data set and mosaic dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import axes
```

```python
HELPrct = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/HELPrct.csv")
```

Let's take a look at the data:

```python
HELPrct.head(1)
```

Use the HELPrct data from the mosaicData to calculate the mean
of all numeric variables (be sure to exclude missing values).

```python
#Calculate the mean of numeric variables
HELPrct.mean()
```

Find the mean of all the numeric variables stratified by sex and age
group where age groups are defined as ranges of 10 years (i.e. 0-10,
10-20, 20-30, etc).

```python
#Create the 10 year age groups and group by sex.
HELPrct["age_groups"] = pd.cut(x=HELPrct["age"], bins=[0, 10, 20, 30, 40, 50, 60])
#Find the mean of all numeric variables:
help = HELPrct.groupby(["sex", "age_groups"]).mean().dropna()
help
```

Using the data set created in the previous problem, create a set of line
plots with the average age of the age group on the x-axis and each of
other numeric variables on the y-axis (in separate plots, obviously!)
stratified by sex.

```python
p = help.groupby("sex").plot.line(subplots=True, figsize = (12,6))

plt.subplots_adjust(wspace=0.3, hspace = 0.3)

plt.show()
```
