This file displays data analysis and data visualizations in Python.

The cubs csv file contains Cubs data between 1876-2022. The variables included are: Year, Team (Tm), League (Lg), Games Played (G), Wins (W), Losses (L), Ties, Winning percentage (W-L%) , Pythagorean winning percentage (pythW-L%), Finish, Number of Fair Ground Balls Hit (GB), Playoffs, Runs (R), Runs Allowed (RA), Attendance, BatAge, PAge, #Bat, #P, Top Player, Managers

Using the cubs csv file, answer the following questions:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
cubs = pd.read_csv("https://raw.githubusercontent.com/gjm112/DSCI401/main/data/cubs_all_time.csv")
```

How many total games have the Cubs won and lost between 1876
and 2022?

```python
sum(cubs.W)
sum(cubs.L)
```

What year did the Cubs score the most runs?

```python
max(cubs.R)
np.argmax(cubs.W)
cubs.Year[np.argmax(cubs.R)]
```

What year did the
Cubs score the fewest runs?

```python
min(cubs.R)
np.argmin(cubs.W)
cubs.Year[np.argmin(cubs.R)]
```

Note that the Cubs had their worst year in 2020, the same year that the COVID-19 pandemic started. Despite their lack of runs, no players had positive test results during the season!

In how many seasons was the Cubs total attendance (i.e. the variable
Attendance) over 3 million?

```python
cubs[cubs['Attendance'] >= 3000000]
sum(cubs.Attendance > 3000000)
```

57.3% of Chicago’s population is Black or Latino. 71% of San
Antonio’s population is Black or Latino. Chicago has given out 301,000
vaccines and 43.4% of these were to Black or Latino residents. San Antonio has given out 266,000 vaccines and 57.2% of these were to Black or
Latino residents. Which of the cities haad a more equitable distribution
of vaccines? (This is an open ended question).

```python
#4. Which city had a more equitable vaccine distribution?

#Data for the bar plots
citypop_BL = np.array([57.3, 71])
citypop_other = np.array([42.7, 29])
vaxed_BL = np.array([43.4, 57.2])
vaxed_other = np.array([56.6, 42.8])
cities = ['Chicago', 'San Antonio']

#Setting the width of the bars and the tick positions
bar_width = 0.35
line_width = 1
pop_bar_pos = np.arange(1,3)
vax_bar_pos = pop_bar_pos + bar_width
tick_pos = [i + (bar_width/50) for i in pop_bar_pos]

#Creating the bars
BL_citypop_bar = plt.bar(pop_bar_pos, citypop_BL, bar_width,
                              color='purple',
                              edgecolor='white',
                              label='Percent of Black or Latino Residents')
other_citypop_bar = plt.bar(pop_bar_pos, citypop_other, bar_width,
                              bottom=citypop_BL,
                              color='pink',
                              edgecolor='white',
                              linewidth=line_width,
                              label='Percent of Non-Black-or-Latino Residents')
BL_vaxed_bar = plt.bar(vax_bar_pos, vaxed_BL, bar_width,
                              color='purple',
                              hatch='//',
                              edgecolor='white',
                              label='Percent of Distributed Vaccines to Black or Latino Residents')
other_vaxed_bar = plt.bar(vax_bar_pos, vaxed_other, bar_width,
                              bottom=vaxed_BL,
                              color="pink",
                              hatch='//',
                              edgecolor='white',
                              linewidth=line_width,
                              label='Percent of Distributed Vaccines to Non-Black-or-Latino Residents')

#Add some texts for labels, titles, and axes ticks
plt.ylabel('Population Percents', fontsize=12)
plt.xlabel('Cities', fontsize=12, labelpad=15)
plt.title('Comparison of Vaccine Distribution Among Ethnic Groups Between Cities', fontsize=20)
plt.xticks(pop_bar_pos + bar_width / 2, cities, rotation=50)

#Create a legend that does not cover the bars
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.show()

#Both cities had about the same equitable distribution of vaccines, they just had different proportions of ethnic groups!
```
